```python
#Step1. matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する。
import matplotlib.pyplot as plt
import numpy as np

#Step2. 6個のグラフを表示する領域を作成します。Figure(描画領域全体)とAxes(ひとつひとつのプロットを描く領域)を作成するコードを記述する。
fig, axes = plt.subplots(2, 3, tight_layout = True)

#Step3. 各領域に、折れ線グラフ、sin関数、ヒストグラム、散布図のデータを表示するコードを記述

##折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

##sin関数
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0, 1].plot(x1, y1, linewidth=2.0)
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10),
               ylim=(-1, 5), yticks=np.arange(-1, 6))

##ヒストグラム
# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x2 = np.random .normal(15, 5, 2000)
axes[0, 2].set(xlim=(-5, 35))
axes[0, 2].hist(x2, bins = 10)

##散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1, 0].scatter(x3, y3)

#Step4. グラフを表示させる行「plt.show()」を書く
plt.show()
```
